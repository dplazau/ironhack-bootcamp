###  Normal Distributions

```python
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
x = np.random.randn(1000) # random numbers that obey a standard normal distribution
```

```python
x
```

```python
sns.distplot(x)
plt.show()
```

```python
normal = norm(5, 2) # initiate normal distribution object
```

```python
x = normal.rvs(1000000)

sns.distplot(x)
```

```python
normal.pdf(5) # the probability of being a zero
```

```python
# Emperical Rule: 68%, 95%, 99.7%
normal.cdf(5) # cumulative function (probability of being 5 or less)
```

```python
normal.rvs(1000) # random value from the normal distribution created
```

```python
x = sorted(normal.rvs(1000))
y = normal.pdf(x)

plt.plot(x,y)
plt.show()
```
